```python
import xml
import xml.etree.ElementTree as ET

m_encoding = 'UTF-8'
```

# HOV Seg 1

```python
hovTau_hovSeg1 = "2.0"
hovMinGap_hovSeg1 = "3.0"
hovSigma_hovSeg1 = "0.2"
hovSpeedFactor_hovSeg1 = "normc(1.0,0.1,1.0,1.2)"
hovCooperative_hovSeg1 = "0.8"
hovStrategic_hovSeg1 = "5"
hovSpeedGain_hovSeg1 = "5"
hovAssertive_hovSeg1 = "1"
hovMaxSpeed_hovSeg1 = "25"

data = ET.Element('additional')

for i, vehLen in enumerate(["5.0", "5.5", "6.0"]):
    ET.SubElement(data, "vType", 
                  id="hov{}_hovSeg1".format(i+1), 
                  vClass="hov",
                  length=vehLen,
                  tau=hovTau_hovSeg1,
                  minGap=hovMinGap_hovSeg1,
                  sigma=hovSigma_hovSeg1,
                  speedFactor=hovSpeedFactor_hovSeg1,
                  lcCooperative=hovCooperative_hovSeg1,
                  lcStrategic=hovStrategic_hovSeg1,
                  lcSpeedGain=hovSpeedGain_hovSeg1,
                  lcAssertive=hovAssertive_hovSeg1,
                  startupDelay="1",
                  maxSpeed=hovMaxSpeed_hovSeg1)
    
    
    ET.SubElement(data, "vType", 
                  id="hov{}_cv_hovSeg1".format(i+1), 
                  vClass="hov",
                  length=vehLen,
                  tau=hovTau_hovSeg1,
                  minGap=hovMinGap_hovSeg1,
                  sigma=hovSigma_hovSeg1,
                  speedFactor=hovSpeedFactor_hovSeg1,
                  lcCooperative=hovCooperative_hovSeg1,
                  lcStrategic=hovStrategic_hovSeg1,
                  lcSpeedGain=hovSpeedGain_hovSeg1,
                  lcAssertive=hovAssertive_hovSeg1,
                  startupDelay="1",
                  maxSpeed=hovMaxSpeed_hovSeg1)
    
    
edge = ["S{}_H".format(i) for i in range(25, 38)]
lane = ["S38_M_4"]


for vType in ["hov1", "hov2", "hov3", "hov1_cv", "hov2_cv", "hov3_cv"]:
    for e in edge:
        calibrator = ET.SubElement(data, "calibrator",
                                   id=e+"_"+vType,
                                   edge=e,
                                   pos="0",
                                   vTypes=vType)
            
        ET.SubElement(calibrator, "flow",
                      begin="0", end=str(7*3600),
                      type=vType+"_hovSeg1")
            
    
    for l in lane:
        calibrator = ET.SubElement(data, "calibrator",
                                   id=l+"_"+vType,
                                   lane=l,
                                   pos="0",
                                   vTypes=vType)
            
        ET.SubElement(calibrator, "flow",
                      begin="0", end=str(7*3600),
                      type=vType+"_hovSeg1")      
            
            
dom = xml.dom.minidom.parseString(ET.tostring(data))
xml_string = dom.toprettyxml()
part1, part2 = xml_string.split('?>')

with open("I210.calibrator(hovSeg1).xml", 'w') as xfile:
    xfile.write(part1 + 'encoding=\"{}\"?>\n'.format(m_encoding) + part2)
    xfile.close()
    
```
